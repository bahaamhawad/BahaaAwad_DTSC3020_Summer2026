## Problem 2 — Sum and Count with a Loop  *(10 points)*

### What to do
1. Use `range()` to create the numbers **1 through 30** (inclusive).
2. With a `for` loop and an **accumulator**, compute:
   - `total_sum` — the sum of all numbers 1–30,
   - `even_count` — how many of them are even (use `%` to test evenness).
3. Inside the loop, print only the **multiples of 5** (5, 10, 15, 20, 25, 30), each on its own line.
4. After the loop, print `total_sum` and `even_count` with labels.

### Hint
`range(1, 31)` gives 1 through 30. A number is even when `number % 2 == 0`.

````python
# Bahaa Awad, Problem 2

# Using range() to create the numbers 1 through 30 (inclusive)
range(1,31)

# Compute with "for" loop and an accumulator
total_sum = 0
even_count = 0

# the sum of all numbers 1-30
for number in range(1,31):
  total_sum += number

  # How many of them are even numbers
  if number % 2 == 0:
    even_count += 1

  # Numbers that are multiple of 5
  if number % 5 == 0:
    print(number)

# Printing with lables
print("Total sum:", total_sum)
print("Even count:", even_count)
````

## Problem 3 — Withdrawal Simulator (while loop)  *(10 points)*

A bank account starts with a balance and you withdraw a fixed amount each round until you can't afford another withdrawal.

### What to do
1. Starter variables: `balance = 100` and `withdrawal = 30`.
2. Use a **`while` loop** that keeps withdrawing `withdrawal` from `balance` **as long as** the balance is at least `withdrawal`.
3. Each round, print the withdrawal and the new balance, e.g. `Withdrew 30, balance is now 70`.
4. When the loop ends, print the **leftover balance** (the amount too small to withdraw) with a label.
5. Add a **comment**: what would happen if `withdrawal` were `0`? (Think about the lecture's warning on infinite loops — one sentence.)
   - **Do not test `withdrawal = 0` by running the loop. Explain it in a comment only.**

### Hint
Your loop condition is `while balance >= withdrawal:`. Make sure something inside the loop changes `balance`, or the loop never ends. If `withdrawal` is `0`, the condition can stay true forever because the balance does not decrease.

````python
balance = 100
withdrawal = 30

# Bahaa Awad, Problem 3

# Starter variables
balance = 100
withdrawal = 30

# Using "while" loop that keeps withdrawing "withdrawal" from "balance" as long as the balance is at least "withdrawal"
while balance >= withdrawal:
  balance -= withdrawal
  print(f"Withdrew {withdrawal}, balance is now {balance}")

print(f"Leftover balance: {balance}")

# Add a Comment if "withdrawal" were 0:
# If withdrawal were 0, the condition 'balance >= withdrawal' would always be true. The balance would never decrease, and this would leading to an infinite loop "Meaning endless scroll of printed results" which may cause the system to freeze or crash.
````

## Problem 4 — Day-Type Router (switch-case style)  *(10 points)*

Convert a day abbreviation into a message, using **either** `match-case` **or** a **dictionary dispatch** (your choice — both are in the lecture).

### Rules
| Code | Message |
|------|---------|
| `"Mon"`, `"Tue"`, `"Wed"`, `"Thu"`, `"Fri"` | `"Weekday — time to study!"` |
| `"Sat"`, `"Sun"` | `"Weekend — take a break!"` |
| anything else | `"Not a valid day code"` |

### What to do
1. Starter variable: `day = "Sat"`.
2. Set a variable `message` using your chosen approach, then print it.
3. **Test every category:** loop a `for` over `["Mon", "Sat", "Fri", "Sun", "Xyz"]` and print the message for each, like `Mon -> Weekday — time to study!`.

### Hint
With `match-case`, you can list several values in one case: `case "Mon" | "Tue" | "Wed" | "Thu" | "Fri":`. With a dictionary, map each code to its message and use `.get(day, "Not a valid day code")`.

````python

# Bahaa Awad, Problem 4

# Starter variable
day = "Sat"

# Set a variable "message" using "dictionary dispatch"
day_messages = {
    "Mon": "Weekday — time to study!",
    "Tue": "Weekday — time to study!",
    "Wed": "Weekday — time to study!",
    "Thu": "Weekday — time to study!",
    "Fri": "Weekday — time to study!",
    "Sat": "Weekend — take a break!",
    "Sun": "Weekend — take a break!"
}

message = day_messages.get(day, "Not a valid day code")
print(f"{day} -> {message}")

# Test every category
test_days = ["Mon", "Sat", "Fri", "Sun", "Xyz"]
for d in test_days:
    test_message = day_messages.get(d, "Not a valid day code")
    print(f"{d} -> {test_message}")

# Note: I tried both match-case and dictionary dispatch.
# I went for the dictionary approach because it felt cleaner for this use case.
````

## Problem 5 — Safe Average Calculator  *(10 points)*

You're given a list of values that *should* be test scores, but some entries are messy. Compute the average of only the valid ones, **without crashing**.

### What to do
The starter list is provided:
```python
raw_values = ["88", "72", "ninety", "95", "", "60", "100"]
```
1. Loop through `raw_values`. Use a `try` / `except ValueError` block to convert each to an `int`.
   - Valid numbers go into a list `valid`.
   - For invalid entries, print a message like `Skipping invalid value: 'ninety'`.
   - For this assignment, a value is **valid** if it can be converted to an integer. You do **not** need to check whether the score is between 0 and 100.
2. After the loop, **check whether `valid` is empty**:
   - If it has numbers, print the count and the average (rounded to 2 decimals).
   - If it's empty, print `No valid values to average.`
3. Add a **comment**: name one Lesson-4 exception (besides `ValueError`) and one situation that would raise it.

### Hint
This mirrors the lecture's *Safe Score Processor*. `int("ninety")` and `int("")` both raise `ValueError`.

````python
raw_values = ["88", "72", "ninety", "95", "", "60", "100"]

# Bahaa Awad, Problem 5

# Looping through "raw_values"
valid = []
for value in raw_values:
    try:
        valid.append(int(value))
    except ValueError:
        print(f"Skipping invalid value: '{value}'")

# Checking whether "valid" is empty
if not valid:
    print("No valid values to average.")
else:
    average = sum(valid) / len(valid)
    print(f"Count: {len(valid)}, Average: {average:.2f}")

# Adding a comment:
# As noted in Lecture 4, some errors occur because an operation is logically impossible, like dividing by zero which raises a ZeroDivisionError:
# print(10 / 0)

# Other errors come from Python's language rules, like data type constraints.
# For example, 'int' stands for integer (numbers), so entering letters raises a ValueError:
# int("hello")
````
