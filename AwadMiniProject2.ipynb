## Part 1 — Drink-Size Price Calculator  *(20 points)*

Eagle Beans charges by drink size, with a small surcharge for oat milk. The owner wants the register to decide the price automatically.

### Price rules
| Size | Base price |
|------|-----------|
| `"small"` | $3.50 |
| `"medium"` | $4.25 |
| `"large"` | $5.00 |

- Oat milk adds **$0.75**, but **only for a valid size**.
- Any size that is **not** `"small"`, `"medium"`, or `"large"` is invalid.
- If the size is invalid, print an error message and set the final price to `0.0`.
- If the size is invalid, **do not add** the oat-milk surcharge, even when `oat_milk` is `True`.

### What to do
1. Start the cell with two comment lines: your name, and a one-line description of what the cell does.
2. Two starter variables are given for you:
   ```python
   drink_size = "medium"
   oat_milk = True
   ```
3. Use an `if` / `elif` / `else` chain to set `base_price` from `drink_size`.
   - In the `else` branch, handle the invalid-size case.
   - You may create a Boolean variable such as `valid_size = True` / `False` if it helps you avoid adding oat milk to an invalid order.
4. Set `final_price` from `base_price`.
5. Use a **second** `if` statement to add the oat-milk surcharge **only when both conditions are true**:
   - `oat_milk` is `True`, and
   - the size is valid.

   You may write the Boolean check as either `if oat_milk == True:` or simply `if oat_milk:`.
6. Print a labeled summary using an f-string with 2 decimals, e.g.:
   ```
   Size: medium | Oat milk: True | Price: $5.00
   ```
7. **Test your logic:** change `drink_size` to `"venti"` and keep `oat_milk = True`. Re-run the cell and confirm:
   - an error message prints,
   - the final price is `$0.00`, and
   - the oat-milk surcharge is **not** added.

   Leave `drink_size` set back to `"medium"` before you submit.

### Hints
- The lecture's grade classifier used the same `if` / `elif` / `else` shape.
- Add the surcharge like the lecture's `+=` examples: `final_price += 0.75`.
- A helpful pattern is: first decide whether the size is valid, then decide whether oat milk should be added.


````python

# Bahaa Awad, Part 1
# Automatically determines drink prices by size, applies an oat milk surcharge, and handles invalid sizes with an error fallback.

#2 Starter variables
drink_size = "medium"
oat_milk = True

# Initialize valid_size as True by default
valid_size = True

#3 Setting the "base_price" from "drink_size" and handling invalid sizes
if drink_size == "small":
    base_price = 3.50
elif drink_size == "medium":
    base_price = 4.25
elif drink_size == "large":
    base_price = 5.00
else:
    valid_size = False
    base_price = 0.0
    print("Error: Invalid drink size entered.")

#4 Setting "final_price" from "base_price"
final_price = base_price

#5 Adding the "oat_milk" surcharge only when both conditions are true
if oat_milk and valid_size:
    final_price += 0.75

#6 Printing a labeled summary using f-string with 2 decimals
print(f"Size: {drink_size} | Oat milk: {oat_milk} | Price: ${final_price:.2f}")

#7 Testing Logic Summary:
# Tested by changing drink_size to "venti" with oat_milk = True.
# Result: Prints an invalid size error message, prevents the oat milk surcharge, and correctly outputs a final price of $0.00.
# Set back to medium for submission
````

## Part 2 — The Morning Rush Loop  *(20 points)*

A whole morning of orders just came in. Instead of adding them one by one, loop through them.

### What to do
The starter list is provided (each number is one order total in dollars):
```python
order_totals = [4.25, 5.00, 3.50, 6.75, 4.25, 8.00, 5.50]
```
1. Use a `for` loop to compute `grand_total`, the sum of every order. **Start an accumulator at 0.0** and add each order inside the loop (the lecture's accumulator pattern — do **not** just call `sum()`).
2. Inside the same loop or a second one, **count** how many orders were **$5.00 or more**. Store it in `big_orders` and print it.
3. Use `range()` **or** `enumerate()` to print a numbered list of the orders, like:
   ```
   Order 1: $4.25
   Order 2: $5.00
   ...
   ```
4. After the loop, print the `grand_total` formatted as money, and print the **average order value** (`grand_total` divided by the number of orders) rounded to 2 decimals.
5. Add a **comment** explaining the difference between `range(len(order_totals))` and `enumerate(order_totals)` in one sentence.

### Hints
- Accumulator: `grand_total = 0.0` before the loop, then `grand_total += order` inside it.
- `enumerate(order_totals, start=1)` numbers the orders starting at 1 (see the lecture's `enumerate()` example).

````python
# Bahaa Awad, Part 2
# Processes a list of morning drink orders to calculate metrics and print a formatted summary.

# Starter variables
order_totals = [4.25, 5.00, 3.50, 6.75, 4.25, 8.00, 5.50]

#1 Initialize accumulators and counters
grand_total = 0.0
big_orders = 0

#2 Loop through orders to calculate grand total, count big orders, and print the numbered list
for index, order in enumerate(order_totals, start=1):
    grand_total += order

    if order >= 5.00:
        big_orders += 1

    print(f"Order {index}: ${order:.2f}")

#3 Print final summaries: grand total, big order, and average order value in dollar and rounded to 2-decimals.
print(f"Grand Total: ${grand_total:.2f}")
print(f"Big Orders ($5.00 or more): {big_orders}")
print(f"Average Order Value: ${grand_total / len(order_totals):.2f}")

#4 Comment explanation:
# range(4len()) provides only the index positions, while enumerate() provides both the index and the corresponding value together.


````

## Part 3 — Loyalty Tier Classifier  *(20 points)*

The owner wants to sort customers into loyalty tiers based on their stars, with a VIP twist.

### Tier rules (use an `elif` ladder)
| Stars | Tier |
|-------|------|
| 50 or more | `"Gold"` |
| 25–49 | `"Silver"` |
| 10–24 | `"Bronze"` |
| fewer than 10 | `"New"` |

### Nested twist
Inside the **Gold** branch only, also check `is_employee`:
- If a Gold customer **is** an employee, set `perk = "Free drinks all week"`.
- If a Gold customer is **not** an employee, set `perk = "One free drink"`.
- For every non-Gold tier, set `perk = "Keep earning stars!"`.

### What to do
1. Starter variables:
   ```python
   loyalty_stars = 32
   is_employee = False
   ```
2. Use an `if` / `elif` / `else` ladder to set `tier`.
3. Use a **nested** `if` (inside the Gold branch) to set `perk` as described.
4. Print both with labels, e.g. `Tier: Silver` and `Perk: Keep earning stars!`.
5. **Test:** set `loyalty_stars = 60` and `is_employee = True`, re-run, confirm you get `Gold` + `Free drinks all week`. Then set the values back to `32` / `False` before submitting.
6. Add a **comment**: why must the `elif` checks be ordered from highest to lowest? (One sentence — what would break if `>= 10` came first?)

````python
# Bahaa Awad, Part 3
# Classifies customers into loyalty tiers based on points, applies custom employee perks within the Gold tier, and displays a summary.

#1 Starter variables
loyalty_stars = 32
is_employee = False

#2 & #3 Setting "tier" and "perk" using an if/elif/else ladder with a nested condition
if loyalty_stars >= 50:
    tier = "Gold"
    if is_employee:
        perk = "Free drinks all week"
    else:
        perk = "One free drink"
elif loyalty_stars >= 25:
    tier = "Silver"
    perk = "Keep earning stars!"
elif loyalty_stars >= 10:
    tier = "Bronze"
    perk = "Keep earning stars!"
else:
    tier = "New"
    perk = "Keep earning stars!"

#4 Printing both with labels
print(f"Tier: {tier}")
print(f"Perk: {perk}")

#5 Test tracking:
# Tested with loyalty_stars = 60 and is_employee = True.
# Result: Tier: Gold | Perk: Free drinks all week.
# After the test and for submission Values have been set back to 32 and False

# #6 Comment explanation:
# If >= 10 came first, anyone with 25 or 50+ stars would evaluate to True immediately and get wrongly trapped in "the Bronze tier which is the tier that matches >= 10" because the ladder stops executing after the first True condition.
# Python evaluate if/elif/else from top to bottom, and it stops as soon as the first condition is true.


````

## Part 4 — Drive-Thru Menu Router  *(20 points)*

### Codes
| Code | Action message |
|------|----------------|
| `"D"` | `"Routing to: Drinks menu"` |
| `"F"` | `"Routing to: Food menu"` |
| `"R"` | `"Routing to: Rewards & loyalty"` |
| `"H"` | `"Routing to: Talk to a human"` |
| anything else | `"Unknown code — please try again"` |

### What to do
1. Starter variable:

   ```python
   menu_code = "R"
   ```
2. Implement the router **either** with `match-case` **or** with a **dictionary dispatch** using `.get(..., default)` — your choice. Both were shown in the lecture.
3. Print the routed action message for the starter value `menu_code = "R"`.
4. **Test all paths:** in the same cell, loop over the list `test_codes = ["D", "F", "R", "H", "Z"]` and print the routed message for each code so the grader can see every branch works:
   ```
   D -> Routing to: Drinks menu

   R -> Routing to: Rewards & loyalty
   H -> Routing to: Talk to a human

   ```
5. Add a **comment** naming one advantage of the approach you used:
   - If you use dictionary dispatch, explain one advantage over a long `if` / `elif` chain.
   - If you use `match-case`, explain one advantage of switch-case style logic over a long `if` / `elif` chain.

### Hints
- `match-case` uses `case _:` for the default (lecture Approach 2).
- Dictionary dispatch: `actions.get(menu_code, "Unknown code — please try again")` (lecture Approach 3).
- The grader should see both the single starter-code output and the five test-code outputs.


````python
# Bahaa Awad, Part 4
# Routes drive-thru single-character user codes to their matching menu actions using structural pattern matching.

#1 Starter variable
menu_code = "R"

#2 & #3 Routing and printing the message for the single starter variable
print("--- Single Code Test ---")
match menu_code:
    case "D":
        print("Routing to: Drinks menu")
    case "F":
        print("Routing to: Food menu")
    case "R":
        print("Routing to: Rewards & loyalty")
    case "H":
        print("Routing to: Talk to a human")
    case _:
        print("Unknown code - please try again")

#4 Testing all paths in a loop to demonstrate every branch works
print("\n--- All Paths Loop Test ---")
test_codes = ["D", "F", "R", "H", "Z"]

for code in test_codes:
    match code:
        case "D":
            print(f"{code} -> Routing to: Drinks menu")
        case "F":
            print(f"{code} -> Routing to: Food menu")
        case "R":
            print(f"{code} -> Routing to: Rewards & loyalty")
        case "H":
            print(f"{code} -> Routing to: Talk to a human")
        case _:
            print(f"{code} -> Unknown code - please try again")

#5 Comment advantage:
# match-case provides a cleaner case-style syntax compared to the more beginner-friendly If-Elif-Else approach.
# I tried using IF-Elif-Else but I had to type menu_code == over and over again for every single condition, therefore, I am going with match-case for cleaner coding.
````

## Part 5 — Crash-Proof Receipt Reader  *(20 points)*

The register sometimes receives messy price data: blanks, typos, words instead of numbers. A real program must **not crash** — it should skip bad data and keep going. This is exactly the lecture's safe-parser pattern.

### What to do
The starter list is provided (these are price strings the register received):
```python
raw_prices = ["4.50", "3.25", "free", "5.00", "", "2.75", "N/A", "6.00"]
```
1. Create two empty lists before the loop:
   ```python
   clean_prices = []
   rejected = []
   ```
2. Loop through `raw_prices`. For each entry, use a `try` / `except` block to convert it to a `float`.
   - On success: add the converted number to `clean_prices`.
   - On failure (`ValueError`): add the original bad entry to `rejected` and print a message like `Could not read price: 'free'`.
3. Use an `else` clause on the inner `try` block. The `else` should run only when conversion succeeded and should print a confirmation like `Accepted: $4.50`.
4. After the loop, print:
   - the `clean_prices` list,
   - the `rejected` list,
   - the **total** of all clean prices, formatted as money.
5. **Guard against an empty list:** if `clean_prices` is empty, print `No valid prices to total.` and do not compute a total. If it is not empty, compute and print the total.
6. Add a `finally` block that prints exactly:
   ```
   Done reading receipt.
   ```
   This message should print **once at the very end**, after all prices have been processed.

   To make this happen, use this structure:
   ```python
   try:
       # create lists
       # loop through raw_prices
       # inside the loop: use try / except / else for each price
       # after the loop: print lists and total
   finally:
       print("Done reading receipt.")
   ```

   Do **not** put the `finally` block inside the loop, because then `Done reading receipt.` would print once for every price instead of once for the whole receipt.

### Hints
- `float("4.50")` works; `float("free")` raises `ValueError` — that's what `except ValueError:` catches.
- Lecture Cell: *Safe Parser for AI-Generated Numeric Data* is the closest model.
- The outer `try` / `finally` handles the whole receipt. The inner `try` / `except` / `else` handles one price at a time.


````python
# Bahaa Awad, Part 5
# Parses messy string receipt data safely using error handling to separate valid floats from bad inputs.

try:
# Starter list
    raw_prices = ["4.50", "3.25", "free", "5.00", "", "2.75", "N/A", "6.00"]

#1 Create empty lists
    clean_prices = []
    rejected = []

#2 Loop through raw prices
    for price in raw_prices:
        try:
            converted_price = float(price)
        except ValueError:
            # Handle conversion failure
            rejected.append(price)
            print(f"Could not read price: '{price}'")
        else:
            #3 Handle conversion success
            clean_prices.append(converted_price)
            print(f"Accepted: ${converted_price:.2f}")

#4 & #5 Post-loop tracking and printing summaries
    print("\n--- Summary ---")
    print(f"Clean Prices: {clean_prices}")
    print(f"Rejected Entries: {rejected}")

    if not clean_prices:
        print("No valid prices to total.")
    else:
        total_price = sum(clean_prices)
        print(f"Total Price: ${total_price:.2f}")

finally:
    #6 Final clean-up block running exactly once at the end
    print("Done reading receipt.")

````
