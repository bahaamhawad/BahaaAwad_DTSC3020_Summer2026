## Part 4: Comments in Python

Comments are notes for humans. Python ignores comments when running the code.

In Python, comments start with the `#` symbol.

```python
# This is a comment.
# Python will not run this line as code.

print("Comments help us explain our code.")
```

## Part 5: Python Statements

A statement is one line of code that tells Python to do something. Python usually runs code line by line, from top to bottom.



```python
print("Hi")
print("Mehri,")
print("Welcome!")
```

## Part 6: Simple Variables

A variable is like a small box that stores a value.

In the example below, `student_name` stores a name, and `course_name` stores the course number.

```python
student_name = "Alex"
course_name = "DTSC 3020"

print(student_name)
print(course_name)
```

We can also combine text and variables in one output.

```python
student_name = "Alex"
course_name = "DTSC 3020"

print("Student name" ,"is:", student_name)
print("Course is:", course_name)
```
