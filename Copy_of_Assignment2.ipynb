## Question 2 — Strings: Indexing, Slicing, and Methods  *(13 points)*

Work with the messy text below and clean it up using the string tools from the lecture.

### What to do
The starter variable is provided in the code cell:

```python
raw_text = "   data science with PYTHON   "
```

1. Print the **length** of `raw_text` using `len(...)`.
2. Create a new variable `clean_text` that has the extra spaces removed (`.strip()`) **and** is all lowercase (`.lower()`). Print it.
3. Print the **first character** and the **last character** of `clean_text` using indexing.
4. Using **slicing**, print just the word `"data"` from `clean_text`.
5. Create a variable `new_text` where the word `"PYTHON"` in `raw_text` is replaced with `"Python 3"` using `.replace(...)`. Print it.
6. Finally, use an **f-string** to print one summary sentence, for example:

   ```
   The cleaned text is 'data science with python' and it has 24 characters.
   ```

   (Use `len(clean_text)` inside the f-string — don't type the number by hand.)

### Hints
- Methods can be chained: `raw_text.strip().lower()`.
- Indexing: `clean_text[0]` is the first character, `clean_text[-1]` is the last.
- Slicing: `clean_text[0:4]` gives the first four characters.
- f-string example: `print(f"It has {len(clean_text)} characters.")`.

````python
raw_text = "   data science with PYTHON   "

# Printing the length
print(len(raw_text))

# Creating a new variable with extra spaces removed and all lowercase
clean_text = raw_text.strip().lower()
print(clean_text)

# Printing the first and last characters
print(clean_text[0])
print(clean_text[-1])

# Printing the word "data", I started from 0 sine the word "data" begins at first character and ended at 4 index. The word "data" is four letters/characters.
print(clean_text[0:4])

# Replacing "PYTHON" with "Python 3"
new_text = raw_text.replace("PYTHON", "Python 3")
print(new_text)

# Printing a summary sentence
print(f"The cleaned text is '{clean_text}' and it has {len(clean_text)} characters.")
````

Use the operator families from the lecture to analyze a student's course results.

### What to do
1. Create these variables:
   - `earned_points = 43`
   - `total_points = 50`
   - `attendance = 0.85`
2. **Arithmetic:** Compute `percentage = earned_points / total_points * 100` and print it. Also print `earned_points // 10` (floor division) and `earned_points % 10` (remainder), each labeled so the output is readable.
3. **Comparison and logical:** Compute and print these booleans:
   - `passed = percentage >= 70 and attendance >= 0.80`
   - `honor  = percentage >= 90 or attendance >= 0.95`
   - `needs_review = not passed`
4. **Assignment operator:** The student earns 5 bonus points. Add them to `earned_points` using `+=`, then print the updated value.
5. **Membership:** A list is provided: `topics = ["variables", "strings", "operators"]`. Print the result of `"strings" in topics` and `"loops" not in topics`.
